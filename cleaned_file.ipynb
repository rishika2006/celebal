```python
df['initial_price'] = pd.to_numeric(
    df['initial_price']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('₹', '', regex=False),
    errors='coerce'
)

df['final_price'] = pd.to_numeric(
    df['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace('â‚¹', '', regex=False),
    errors='coerce'
)
```
