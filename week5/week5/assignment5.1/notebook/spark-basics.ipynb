```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder \
    .appName("Spark Assignment") \
    .getOrCreate()
```

```python
df = spark.read \
    .option("header", True) \
    .option("inferSchema", True) \
    .option("multiLine", True) \
    .option("quote", '"') \
    .option("escape", '"') \
    .csv("../data/Sample - Superstore.csv")

df.show(5)
```

```python
df.printSchema()
```

```python
df.filter(
    col("Sales") > 500
).show(5)
```

```python
df.filter(
    col("Region") == "West"
).show(5)
```

```python
df.filter(
    col("Region") == "West"
).show(5)
```

```python
df.filter(
    col("Category") == "Technology"
).show(5)
```

```python
df.select(
    [count(when(col(c).isNull(), c)).alias(c)
     for c in df.columns]
).show()
```

```python
df = df.withColumnRenamed(
    "Sub-Category",
    "SubCategory"
)
```

```python
df = df.withColumn(
    "DiscountAmount",
    col("Sales") * col("Discount")
)

df.show(5)
```

```python
df.select(
    count("*").alias("Total Orders"),
    avg("Sales").alias("Average Sales"),
    min("Profit").alias("Minimum Profit"),
    max("Profit").alias("Maximum Profit")
).show()
```

```python
df.groupBy("Category").agg(
    count("*").alias("Orders"),
    sum("Sales").alias("Total Sales"),
    avg("Profit").alias("Average Profit")
).show()
```

```python
df.groupBy("Region").agg(

    sum("Sales").alias("Total Sales")

).show()
```

```python
df.groupBy("State") \
.agg(
sum("Sales").alias("Total Sales")
) \
.orderBy(
col("Total Sales").desc()
).show(5)
```

```python
pipeline = (

df.dropDuplicates()

.fillna("Unknown")

.filter(col("Sales") > 100)

.groupBy("Category")

.agg(
sum("Sales").alias("Total Sales")
)

)

pipeline.show()
```

Top 10 products by sales


```python
df.groupBy("Product Name") \
  .agg(sum("Sales").alias("Total Sales")) \
  .orderBy(col("Total Sales").desc()) \
  .show(10)
```

Category-wise profit

```python
df.groupBy("Category") \
  .agg(sum("Profit").alias("Total Profit")) \
  .show()
```

Region-wise average sales

```python
df.groupBy("Region") \
  .agg(avg("Sales").alias("Average Sales")) \
  .show()
```

Find loss-making orders

```python
df.filter(
    col("Profit") < 0
).show(10)
```

Top 5 customers by sales

```python
df.groupBy("Customer Name") \
  .agg(sum("Sales").alias("Total Sales")) \
  .orderBy(col("Total Sales").desc()) \
  .show(5)
```

Create a Profit Category column

```python
from pyspark.sql.functions import when

df = df.withColumn(
    "ProfitStatus",
    when(col("Profit") > 0, "Profit")
    .otherwise("Loss")
)

df.select(
    "Product Name",
    "Profit",
    "ProfitStatus"
).show(10)
```
