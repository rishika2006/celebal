# Spark Assignment
Objective

Understand Spark architecture and perform efficient data processing using PySpark.

```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Spark Assignment") \
    .getOrCreate()

spark
```

```python
df = spark.read.csv(
    "/employees.csv",
    header=True,
    inferSchema=True
)

df.show()
```

# Print Schema
The schema provides information about the column names and their data types. It helps us verify that Spark has correctly inferred the structure of the dataset.

```python
df.printSchema()
```

# Display Dataset
The show() function displays the contents of the DataFrame for verification.



```python
df.show()
```

# Selecting Required Columns
Selecting only the required columns improves performance by reducing unnecessary data processing.

```python
df.select("Name", "Department", "Salary").show()
```

# Filtering Records
Filter employees whose salary is greater than 50000.

```python
df.filter(df.Salary > 50000).show()
```

# Renaming Column
Rename the Salary column to Monthly_Salary for better readability.

```python
df = df.withColumnRenamed("Salary", "Monthly_Salary")

df.show()
```

# Changing Data Type
The Monthly_Salary column is converted to Integer type to ensure consistency and enable arithmetic operations.

```python
from pyspark.sql.functions import col

df = df.withColumn(
    "Monthly_Salary",
    col("Monthly_Salary").cast("int")
)

df.printSchema()
```

# Adding New Column
Create a new column Annual_Salary by multiplying the monthly salary by 12

```python
df = df.withColumn(
    "Annual_Salary",
    col("Monthly_Salary") * 12
)

df.show()
```

# Handling NULL Values

```python
df = df.fillna({
    "Monthly_Salary": 0
})

df.show()
```








# Filtering Employees
Display employees whose monthly salary is greater than 50000.

```python
df.filter(col("Monthly_Salary") > 50000).show()
```

# Data Processing Pipeline

```python
pipeline = (
    df
    .filter(col("Age") > 24)
    .select(
        "Name",
        "Department",
        "Monthly_Salary",
        "Annual_Salary"
    )
)

pipeline.show()
```

# Fixing null value of Neha


```python
df = df.fillna({
    "Monthly_Salary": 0
})

df = df.withColumn(
    "Annual_Salary",
    col("Monthly_Salary") * 12
)

df.show()
```

# Saving Data as CSV

```python
pipeline.write.mode("overwrite") \
    .option("header", True) \
    .csv("output/csv_output")
```

## Saving Data as Parquet

The transformed dataset is also saved in Parquet format, which is optimized for Spark processing.

```python
pipeline.write.mode("overwrite") \
    .parquet("output/parquet_output")
```

## Reading Parquet File

```python
parquet_df = spark.read.parquet("output/parquet_output")

parquet_df.show()
```

## Execution Plan

The execution plan helps understand how Spark optimizes transformations before execution.

```python
pipeline.explain(True)
```
